# Cost curve plots

```python
%pylab inline
import numpy as np
from matplotlib import cm
```

```python
ax = plt.subplot(111)
plt.xlim(0,10)
plt.ylim(0,10)
plt.yticks([])
plt.xticks([])
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# suppression line
ax.plot([0,3],[8,0], 'k', linewidth=3)

# constant cost lines
colors = [ cm.autumn(x) for x in np.linspace(1,0,10) ]
cratio = 1.5
for c, color in zip(range(1,10), colors):
    ax.plot([0, cratio*c],[c,0], linewidth=1, color=color)

plt.legend(['Suppression Line'],frameon=False)
plt.xlabel('Repressive Threat', fontsize=14)
plt.ylabel('Concessions', fontsize=14)
```

```python
plt.figure(figsize=(10,4))
ax = plt.subplot(121)
plt.xlim(0,10)
plt.ylim(0,10)
plt.yticks([])
plt.xticks([])
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# suppression line
ax.plot([0,7],[3,0], 'k', linewidth=3)
   
# constant cost lines    
xs, ys = np.meshgrid(np.linspace(0,10,100), np.linspace(0,10,100))
zs = xs + ys
CS = plt.contour(xs,  ys, zs, levels=np.arange(0,20,1), cmap='autumn_r')    
plt.clabel(CS, inline=1, fontsize=10, fmt='%i')

plt.legend(['Suppression Line'],frameon=False)
plt.xlabel('Repressive Threat', fontsize=14)
plt.ylabel('Concessions', fontsize=14)
plt.title('High Probability of Success', fontsize=16)

#########################
ax = plt.subplot(122)
plt.xlim(0,10)
plt.ylim(0,10)
plt.yticks([])
plt.xticks([])
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# suppression line
ax.plot([0,4],[8,0], 'k', linewidth=3)

# constant cost lines
xs, ys = np.meshgrid(np.linspace(0,10,100), np.linspace(0,10,100))
zs = xs + ys
CS = plt.contour(xs,  ys, zs, levels=np.arange(0,20,1), cmap='autumn_r')   
plt.clabel(CS, inline=1, fontsize=10, fmt='%i')

plt.legend(['Suppression Line'],frameon=False)
plt.xlabel('Repressive Threat', fontsize=14)
plt.ylabel('Concessions', fontsize=14)
plt.title('Low Probability of Success', fontsize=16)
```
